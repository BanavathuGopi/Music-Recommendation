```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
path='C:Copy the path of csv file here'
song_data= pd.read_csv(path)
song_data.head(10)
```

```python
df = song_data.sample(5000).reset_index(drop=True)
df.shape
```

```python
selected_features = ['title','artist_name']
print(selected_features)
```

```python
for f in selected_features:
    df[f]=df[f].fillna('')
```

```python
combine_features = df['title']+''+df['artist_name']
combine_features = combine_features.fillna('')
print(combine_features)
```

```python
vectorizer = TfidfVectorizer()
f_vectors = vectorizer.fit_transform(combine_features)
print(f_vectors)


# tfidf_vectorizer = TfidfVectorizer()
# tfidf_matrix = tfidf_vectorizer.fit_transform(combine_features)
# tfidf_matrix_csr = csr_matrix(tfidf_matrix)
# print(tfidf_matrix_csr)
```

```python
similarity = cosine_similarity(f_vectors)
print(similarity)

# n_components = 3
# svd = TruncatedSVD(n_components=n_components)
# tfidf_matrix_svd = svd.fit_transform(tfidf_matrix)

# similarity_svd = cosine_similarity(tfidf_matrix_svd)
```

```python
def recomandation(title):
    idx=df[df['title'] == title].index[0]
    d= sorted(list(enumerate(similarity[idx])),reverse=False,key=lambda x:x[1])
    title = []
    for i in d[1:21]:
        title.append(df.iloc[i[0]].title)
    return title
```
